```python
import os
import subprocess
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt

def is_path(path):
  return os.path.exists(path)

def unzip_file(src_path,dst_path,keep_folders=False):
  keep_msg='-j' if not keep_folders else ''
  if not is_path(src_path): return print('src_path does not exist!')
  subprocess.run(['unzip',keep_msg,src_path,'-d',dst_path])

# exclude rightend, right in interval, i.e. [left,right)
def resize_interval(leftend,rightend,left,right,size):
  max_size=rightend-leftend
  curr_size=right-left
  if size==curr_size: return left,right,size
  elif size<curr_size:
    left+=random.randint(0,curr_size-size)
    right=left+size
    return left,right,size
  else: # size>curr_size
    if size>=max_size:
      return leftend,rightend,max_size
    else: # size<max_size
      diff=size-curr_size
      left_space=left-leftend
      right_space=rightend-right
      left-=random.randint(max(0,diff-right_space),min(diff,left_space))
      right=left+size
      return left,right,size

def crop_helper(img_w,img_h,x,y,crop_w,crop_h,size=256):
  size=min(size,img_w,img_h)
  x,_,crop_w=resize_interval(0,img_w,x,x+crop_w,size)
  y,_,crop_h=resize_interval(0,img_h,y,y+crop_h,size)
  return x,y,size

def crop(img,x,y,crop_w,crop_h,size=256):
  img_h,img_w,_=img.shape
  x,y,crop_size=crop_helper(img_w,img_h,x,y,crop_w,crop_h,size)
  img=img[y:y+crop_size,x:x+crop_size]
  img=cv2.resize(img,(size,size),cv2.INTER_AREA)
  return img

def plot(nrows,ncols,imgs,figsize=8):
  total=len(imgs)
  fig=plt.figure(figsize=(figsize,figsize))
  for i in range(1, min(ncols*nrows,total)+1):
    img=imgs[i-1]
    fig.add_subplot(nrows,ncols,i)
    plt.imshow(img)
  plt.show()
```
